```python
#########################
### Umwandlung in wav ###
#########################

# Zuerst FFmpeg installieren
!apt-get install -y ffmpeg

# Importiere benötigte Bibliotheken
import os

# Definiere die Dateinamen und Ausgabe-Pfade
input_files = ["/content/te_hs25_session01_Intro.mp4"
                ]

output_files = [file.replace(".mp4", ".wav") for file in input_files]

# Konvertiere jede Datei in eine WAV-Datei
for input_file, output_file in zip(input_files, output_files):
    !ffmpeg -i "{input_file}" -ac 1 -ar 16000 -sample_fmt s16 "{output_file}"
    print(f"Konvertiert: {input_file} -> {output_file}")

# Überprüfe, ob die Dateien korrekt erstellt wurden
for output_file in output_files:
    if os.path.exists(output_file):
        print(f"Datei {output_file} wurde erfolgreich erstellt.")
    else:
        print(f"Fehler beim Erstellen der Datei {output_file}.")


#####################
### Transkription ###
#####################

# Installiere benötigte Bibliotheken
!pip install openai-whisper python-docx

# Importiere die Bibliotheken
import whisper
from docx import Document

# Lade das Whisper Model (small/medium/large)
model = whisper.load_model("medium")

# Definiere die Dateinamen und Pfade
input_files = ["/content/te_hs25_session01_Intro.wav"
                ]

# Funktion zur Transkription und Speicherung als Word-Dokument
def transcribe_and_save(input_file):
    # Transkription
    result = model.transcribe(input_file)
    text = result["text"]

    # Erstelle ein Word-Dokument
    doc = Document()
    doc.add_heading(f'Transkription von {input_file}', 0)
    doc.add_paragraph(text)

    # Speichere das Dokument
    output_file = input_file.replace(".wav", ".docx")
    doc.save(output_file)
    print(f"Transkript für {input_file} gespeichert als {output_file}")

# Verarbeite jede Datei
for file in input_files:
    transcribe_and_save(file)
```
